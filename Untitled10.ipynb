```python
import pandas as pd
```

```python
df=pd.read_csv('/content/titanic (2).csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
print(df.isnull().sum())
```

```python
# Select only numerical features for correlation analysis
numerical_features = df.select_dtypes(include=['number'])
correlation_matrix = numerical_features.corr()

# Display the correlation matrix using a heatmap
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Fare', data=df, hue='Survived', # Added data=df
                palette='Set1')
plt.title('Age vs Fare: Trend Analysis')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=30, kde=True, color='skyblue') # Changed data to df
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Fare'], color='lightcoral')
```

```python
plt.title('Fare Outliers')
plt.xlabel('Fare')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', data=df, palette='Set2')
plt.title('Passenger Class Distribution')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', data=df, palette='Set2')
plt.title('Survival Distribution')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```
